```python
#Importar librería de pandas para el manejo de los datos
#Juan Carlos Solorzano Castanedo - Big Data - 28/Marzo/2023
import pandas as pd
```

```python
#importar los datos del mapa de siniestralidad como un DataFrame de pandas y mostrar la información
df = pd.read_csv("../data/mapa_siniestralidad.csv")
df
```

```python
#Filtrar los accidentes que involucren a las Bicicletas 
accidentes_bici = df[df["ibaen_atro"] == "Bicicleta"]
accidentes_bici
```

```python
accidentes_bici = accidentes_bici.dropna()
accidentes_bici
```

```python
#Guardar el archivo del DataFrame resultante
accidentes_bici.to_csv('../data/mapa_siniestralidad_bicicletas.csv', index="false")
```
