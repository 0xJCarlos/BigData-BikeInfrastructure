```python
import folium
import pandas as pd
from folium.plugins import MarkerCluster, FeatureGroupSubGroup, HeatMap
```

```python
datos_bicicletas = pd.read_csv("../data/mapa_siniestralidad_bicicletas.csv")
datos_semaforos = pd.read_excel("../data/coordenadas_semaforos_amim.xlsx")
```

```python
datos_bicicletas
```

```python
datos_bicicletas.rename(columns={'y':'latitude','x':'longitude'},inplace=True)
```

```python
locations1 = datos_bicicletas[['latitude','longitude']]
locations2 = datos_semaforos[['latitude','longitude']]
```

```python
data3 = pd.merge(locations1, locations2, on=['latitude', 'longitude'])
```

```python
data3
```

```python
map1_data = folium.FeatureGroup(name='Accidentes de Bicicletas')
map2_data = folium.FeatureGroup(name='Reportes de Semaforos')
```

```python
marker_cluster1 = MarkerCluster().add_to(map1_data)
marker_cluster2 = MarkerCluster().add_to(map2_data)
```

```python
gradient1 = {0.4: '#f6fff3', 0.65: '#006fd8', 1: '#191c8c'}
gradient2 = {0.4: '#fdfdef', 0.65: '#eb0000', 1: '#c70000'}
```

```python
heatmap1 = HeatMap(locations1, gradient=gradient1, radius=20, blur=30, max_zoom=10, max_opacity=0.5)
heatmap2 = HeatMap(locations2, gradient=gradient2, radius=20, blur=30, max_zoom=10, max_opacity=0.5)
```

```python
marker_cluster1.add_child(heatmap1)
marker_cluster2.add_child(heatmap2)
```

```python
base_map = folium.Map(location=[20.6720375, -103.338396], zoom_start=12, tiles='cartodbdark_matter')
```

```python
layer1 = folium.plugins.FeatureGroupSubGroup(base_map, 'Accidentes de Bicicletas')
layer2 = folium.plugins.FeatureGroupSubGroup(base_map, 'Reportes de Semaforos')
```

```python
layer1.add_child(marker_cluster1)
layer2.add_child(marker_cluster2)
```

```python
base_map.add_child(layer1)
base_map.add_child(layer2)
```

```python
layer_control = folium.LayerControl()
base_map.add_child(layer_control)
```
